1.First Dataset ----

```python
import pandas as pd
import numpy as np

data=pd.DataFrame({
    'age':[30,2,12,4,32,33,69],'color':['blue','green','red','white','gray','black','red'],'food':['steak','lamb','mango','apple','cheese','melon','beans'],'height':[165,70,120,80,180,172,150],'score':[4.6,8.3,9.0,3.3,1.8,9.5,2.2],'state':['NY','TX','FL','AL','AK','TX','TX']},index=["Jane","Nick","Aron","Penelope","Dean","Christina","Cornelia"])
data
```

```python
# Selecting column Penelope

display(data.loc[['Penelope']])
```

```python
#columns from Aron to Dean

display(data.loc['Aron':'Dean'])
```

```python
# selecting columns jane,dean,cornelia
display(data.loc[["Jane","Dean","Cornelia"]])
```

```python
#selecting 4th index column
display(data.iloc[[4]])
```

```python
data2 = pd.DataFrame({'Brand' : ['Maruti', 'Hyundai', 'Tata', 'Mahindra', 'Maruti', 'Hyundai', 'Renault', 'Tata', 'Maruti'], 'Year' : [2012, 2014, 2011, 2015, 2012,
2016, 2014, 2018, 2019], 'Kms Driven' : [50000, 30000, 60000,
25000, 10000, 46000,
31000, 15000, 12000], 'City' : ['Gurgaon', 'Delhi', 'Mumbai',
'Delhi', 'Mumbai', 'Delhi',
'Mumbai','Chennai', 'Ghaziabad'], 'Mileage' : [28, 27, 25, 26, 28,
29, 24, 21, 24]})

data2
```
